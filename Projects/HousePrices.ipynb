```python
cat_col_train = ['FireplaceQu','GarageType','GarageFinish','MasVnrType','BsmtQual',
           'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu',
          'GarageQual','GarageCond']

num_col_train = ['LotFrontage','GarageYrBlt','MasVnrArea']

for i in cat_col_train:
    Train[i]=Train[i].fillna(Train[i].mode()[0])

for i in num_col_train:
    Train[i]=Train[i].fillna(Train[i].median())
```

```python
cat_col_test=['FireplaceQu','GarageType','GarageFinish','MasVnrType','BsmtQual',
           'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu',
          'GarageQual','GarageCond','MSZoning','Utilities','Exterior1st','Exterior2nd','KitchenQual','Functional','SaleType']

num_col_test=['LotFrontage','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageYrBlt','GarageCars','GarageArea']
```

```python
for i in cat_col_test:
    Test[i]=Test[i].fillna(Test[i].mode()[0])

for i in num_col_test:
    Test[i]=Test[i].fillna(Test[i].median())

    
```

```python
drops=['Id','Alley','PoolQC','Fence','MiscFeature']

for k in drops:
    Train.drop(columns=k,axis=1,inplace=True)
    Test.drop(columns=k,axis=1,inplace=True)

```

```python
print(np.shape(Test))
print(np.shape(Train))
```

```python
df= pd.concat([Test,Train],axis=0)
df.head()
```

```python
df.columns
```

```python
df["SalePrice"]=df["SalePrice"].fillna(value=df['SalePrice'].mean())
df.isnull().sum()
```

```python

saleprice=df["SalePrice"]
df=df.drop(columns="SalePrice")

numeric_features= df.dtypes[df.dtypes!='object'].index
df[numeric_features]=df[numeric_features].apply(lambda x: (x-x.mean())/(x.std()))
df=pd.get_dummies(df,dummy_na=True).astype(float)
df=pd.concat([df,saleprice],axis=1)
df
```

```python
train=df[1459:][:]
test=df[:1459][:]
print(np.shape(train))
print(np.shape(test))
```

```python
X_train=train[train.columns[:-1]].values
y_train=train[train.columns[-1]].values

X_test=test[test.columns[:-1]].values
y_test=test[test.columns[-1]].values
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
pred_dt = dt_model.predict(X_test)
print(pred_dt.shape)
```

```python
sub_df = pd.read_csv('Dataset/HousePrices/sample_submission.csv')
sub_df['SalePrice'] = pred_dt
sub_df.to_csv('sample_sub_dt.csv', index = False)
```
